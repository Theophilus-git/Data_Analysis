## Data Visualization with seaborn


Seaborn is a python visualization library based on Matplotlib that provide a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with a few lines of code.

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
## Basic Plotting with Seaborn
data = sns.load_dataset('tips')
data
```

```python
# Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x = 'total_bill', y = 'tip',data=data)
plt.title("Scatter plot of Total Bills vs Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=data)
plt.title("Line Plot of total bill by size")
plt.show()
```

```python
# Categorical Plots
# Bar plot

sns.barplot(x='day',y='total_bill',data=data)
plt.title('Bar plot of Total Bill by Day')
plt.show()
```

```python
# Box Plot

sns.boxplot(x='day',y='total_bill',data=data)
plt.show()
```

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=data)
```

```python
## Histograms

sns.histplot(data['total_bill'],bins=10,kde=True)
```

```python
# KDE Plot

sns.kdeplot(data['total_bill'],shade=True)
```

```python
# Pairplot

sns.pairplot(data)
```

```python
## HEatmap

corr=data[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
